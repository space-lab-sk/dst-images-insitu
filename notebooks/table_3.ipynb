## Table 3 reproduction

Table 3 can be reproduced by running [gru-ie.py](../experiments/src/gru-ie.py) and [gru-attn-iec.py](../experiments/src/gru-attn-iec.py) multiple times. 2 models * 5 seeds * 5 folds * 2 forecast horizons = 100 runs.   
For more effecient training process, we can call trainings from bash script in [run_experiments.sh](../experiments/run_experiments.sh).   
You can write there same command as you would run in terminal e.g.:

```sh
python src/gru-ie.py -cfn=conf_gru_ie_1_1_1.yaml -dev=cpu -s=1234
```

or

```sh
python src/gru-attn-iec.py -cfn=conf_gru_attn_iec_1_1_1.yaml -dev=cuda:0 -s=1234
```

-s is chosen seed and -dev is device, either 'cpu' or 'cuda', if more GPUs available you can specify which GPU to use with :0 or :1 etc.
-cfn parameter is for configuration filename. You can name it as you wish, but we sticked with following convention:

"approachName_experimentalBatchNum_differentHyperNum_foldNum.yaml"

approachName - gru_ie or gru_attn_iec
experimentalBatchNum - we had multiple experimental runs which we ran together
differentHyperNum - when we experimented with different hyperparameter setting, we distingiushed configs with this number
foldNum - numer which k-fold was model trained, validated and tested on.

So to run for example "gru_ie" approach for table 3 final results:

```sh
python src/gru-ie.py -cfn=conf_gru_ie_1_1_1.yaml -dev=cpu -s=42
python src/gru-ie.py -cfn=conf_gru_ie_1_1_1.yaml -dev=cpu -s=71
python src/gru-ie.py -cfn=conf_gru_ie_1_1_1.yaml -dev=cpu -s=143
python src/gru-ie.py -cfn=conf_gru_ie_1_1_1.yaml -dev=cpu -s=176
python src/gru-ie.py -cfn=conf_gru_ie_1_1_1.yaml -dev=cpu -s=821

python src/gru-ie.py -cfn=conf_gru_ie_1_1_2.yaml -dev=cpu -s=42
python src/gru-ie.py -cfn=conf_gru_ie_1_1_2.yaml -dev=cpu -s=71
python src/gru-ie.py -cfn=conf_gru_ie_1_1_2.yaml -dev=cpu -s=143
python src/gru-ie.py -cfn=conf_gru_ie_1_1_2.yaml -dev=cpu -s=176
python src/gru-ie.py -cfn=conf_gru_ie_1_1_2.yaml -dev=cpu -s=821

# etc.... 1_1_3, 1_1_4, 1_1_5, 1_2_1, 1_2_2, 1_2_3, 1_2_4, ....until:

python src/gru-ie.py -cfn=conf_gru_ie_1_2_5.yaml -dev=cpu -s=42
python src/gru-ie.py -cfn=conf_gru_ie_1_2_5.yaml -dev=cpu -s=71
python src/gru-ie.py -cfn=conf_gru_ie_1_2_5.yaml -dev=cpu -s=143
python src/gru-ie.py -cfn=conf_gru_ie_1_2_5.yaml -dev=cpu -s=176
python src/gru-ie.py -cfn=conf_gru_ie_1_2_5.yaml -dev=cpu -s=821

```
